```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)

# Explore the data
print(data.head())
```

```python
X = data.iloc[:, :-1].values  # Independent variable (Study Hours)
y = data.iloc[:, 1].values    # Dependent variable (Percentage Score)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the test set results
y_pred = regressor.predict(X_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, regressor.predict(X_train), color='red')
plt.title('Percentage vs Study Hours (Training set)')
plt.xlabel('Study Hours')
plt.ylabel('Percentage Score')
plt.show()
```

```python
hours = 9.25
predicted_score = regressor.predict([[hours]])
print("Predicted Score for {} hours of study: {:.2f}".format(hours, predicted_score[0]))
```
